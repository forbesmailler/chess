```python
from sklearn.feature_selection import mutual_info_classif, f_classif
import pandas as pd
```

```python
X = pd.read_parquet("X_train.parquet")
y = pd.read_parquet("y_train.parquet")["value"]

X = X.iloc[:, :768]
X = (X != 0).astype(int)
```

```python
X.sum(axis=0).sort_values().head(50)
```

```python
# 1a) ANOVA F-score (treating –1,0,1 as three classes)
F, p = f_classif(X, y)
uni = pd.DataFrame({
    'feature': X.columns,
    'F_score': F,
    'p_value': p
}).sort_values('F_score', ascending=False)

# 1b) Mutual information
mi = mutual_info_classif(X, y, discrete_features=True)
uni['MI'] = mi
uni = uni.sort_values('MI', ascending=False)

# Peek at the top 20
print(uni.head(20))
```
